```python
page = requests.get('https://www.imdb.com/list/ls056092300/')
```

```python
page
```

```python
Soup=BeautifulSoup(page.content)
```

```python
Soup
```

```python
titles=[]
movie_items = Soup.find_all('div', class_='lister-item-content')

for item in movie_items:
    titles.append(item.h3.a.text)

print(titles)
```

```python
ratings=[]
movie_items = Soup.find_all('div', class_='lister-item-content')

for item in movie_items:
    rating = item.find('span', class_='ipl-rating-star__rating').text
    ratings.append(float(rating))
print(ratings)
```

```python
years=[]
movie_items = Soup.find_all('div', class_='lister-item-content')

for item in movie_items:
    year = item.find('span', class_='lister-item-year').text
    years.append(year)
print(years)   
```

```python
df=pd.DataFrame({'Name':titles,'Rating':ratings,'Year':years})
print(df)
```
